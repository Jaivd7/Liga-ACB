```python
#Acessing the website
import pandas as pd
import requests
from bs4 import BeautifulSoup



website_url = requests.get('https://www.basketball-reference.com/international/spain-liga-acb/2015.html').text
soup = BeautifulSoup(website_url,'lxml')

website_url2 = requests.get('https://www.basketball-reference.com/international/spain-liga-acb/2016.html').text
soup2 = BeautifulSoup(website_url2,'lxml')

website_url3 = requests.get('https://www.basketball-reference.com/international/spain-liga-acb/2017.html').text
soup3 = BeautifulSoup(website_url3,'lxml')

website_url4 = requests.get('https://www.basketball-reference.com/international/spain-liga-acb/2018.html').text
soup4 = BeautifulSoup(website_url4,'lxml')

website_url5 = requests.get('https://www.basketball-reference.com/international/spain-liga-acb/2019.html').text
soup5 = BeautifulSoup(website_url5,'lxml')

website_url6 = requests.get('https://www.basketball-reference.com/international/spain-liga-acb/2020.html').text
soup6 = BeautifulSoup(website_url6,'lxml')


```

```python
#Acessing the table

table = soup.find('table',{'class':'sortable stats_table'})
table2 = soup2.find('table',{'class':'sortable stats_table'})
table3 = soup3.find('table',{'class':'sortable stats_table'})
table4 = soup4.find('table',{'class':'sortable stats_table'})
table5 = soup5.find('table',{'class':'sortable stats_table'})
table6 = soup6.find('table',{'class':'sortable stats_table'})
```
